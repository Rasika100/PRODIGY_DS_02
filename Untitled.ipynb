```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_df = sns.load_dataset('titanic')
```

```python
print(titanic_df.head())
```

```python
print(titanic_df.describe())
```

```python
print(titanic_df.isnull().sum())
```

```python
titanic_df['age'] = titanic_df['age'].fillna(titanic_df['age'].median())
```

```python
titanic_df['fare']=titanic_df['fare'].fillna(titanic_df['fare'].median())
```

```python
titanic_df['embarked_Q'] = titanic_df['embarked_Q'].fillna(titanic_df['embarked_Q'].mode()[0])

# For 'embarked_S', fill missing values with mode
titanic_df['embarked_S'] = titanic_df['embarked_S'].fillna(titanic_df['embarked_S'].mode()[0])
```

```python
print(titanic_df.isnull().sum())
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.histplot(titanic_df['age'], bins=20, kde=True)
plt.title('Age Distribution')
```

```python
plt.subplot(2, 2, 2)
sns.histplot(titanic_df['fare'], bins=20, kde=True)
plt.title('Fare Distribution')
```

```python
plt.subplot(2, 2, 3)
sns.countplot(data=titanic_df, x='survived')
plt.title('Survival Count')
```

```python
plt.subplot(2, 2, 4)
sns.countplot(data=titanic_df, x='pclass')
plt.title('Passenger Class Count')
```

```python
plt.tight_layout()
plt.show()
```

```python
print(titanic_df.dtypes)
```

```python
numeric_columns = titanic_df.select_dtypes(include=['int64', 'float64', 'bool'])
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.barplot(data=titanic_df, x='pclass', y='survived')
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
sns.barplot(data=titanic_df, x='sex_male', y='survived')
plt.title('Survival Rate by Sex')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_df, x='age', hue='survived', kde=True)
plt.title('Survival Rate by Age')
plt.show()
```
